This notebook takes an omnibus FCC download of license data and parses it into a dictionary of dataframes described in the schemas provided at http://wireless.fcc.gov/uls/documentation/pa_ddef42.pdf

```python
import pandas as pd
```

```python
# The file downloaded from fcc.gov ULS site:
# http://wireless2.fcc.gov/UlsApp/UlsSearch/searchAdvanced.jsp

f = './UL20169142344305.txt'
```

```python
schemafile = './pa_ddef42.xls'
```

```python
schemas_xl = pd.read_excel(schemafile, sheetname=None, skiprows=1)
```

```python
alltables = pd.read_csv(f,sep='|', header=None, names=range(1,100))
```

```python
tnames = [x for x in alltables[1].unique() if len(x)==2]
```

```python
alltables[1].unique()
```

```python
tnames
```

```python
tables = {x: alltables[alltables[1]==x] for x in tnames}
```

```python
LO = tables['LO'].dropna(axis=1, how='all')
```

```python
schemas = {x: schemas_xl[x].set_index('Position').T.drop('Definition').to_dict('records')[0] for x in tnames}
```

```python
schemas['MW']
```

```python
tables = {x: tables[x].rename(columns=schemas[x]).dropna(axis=1, how='all') for x in tnames}
```
